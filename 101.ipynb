```
import pandas as pd
import tensorflow as tf
```

```
파일경로 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
iris = pd.read_csv(파일경로)
# 원핫인코딩
iris = pd.get_dummies(iris)
iris
```

```
iris.columns
```

```
독립 = iris[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
종속 = iris[['품종_setosa', '품종_versicolor', '품종_virginica']]
print(독립.shape, 종속.shape)
```

```
X = tf.keras.layers.Input(shape=[4])
Y = tf.keras.layers.Dense(3, activation='softmax')(X)   # sigmoid, softmax(확률!) 차이점!!(분류) <-> Identity(회귀)
model = tf.keras.models.Model(X,Y)
model.compile(loss='categorical_crossentropy', metrics='accuracy')    # categorical_crossentropy(분류) <-> mse(회귀)
```

```
model.fit(독립, 종속, epochs=100)
```

```
model.predict(독립[-5:])
```

```
파일경로 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
iris = pd.read_csv(파일경로)
iris = pd.get_dummies(iris)
iris.columns
```

```
독립 = iris[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
종속 = iris[['품종_setosa', '품종_versicolor', '품종_virginica']]
```

```
X = tf.keras.layers.Input(shape=[4])
H = tf.keras.layers.Dense(8, activation='swish')(X)
H = tf.keras.layers.Dense(8, activation='swish')(H)
H = tf.keras.layers.Dense(8, activation='swish')(H)
Y = tf.keras.layers.Dense(3, activation='softmax')(H)
model = tf.keras.models.Model(X, Y)
model.compile(loss='categorical_crossentropy', metrics='accuracy')
model.summary()
```

```
model.fit(독립, 종속, epochs=100)
```

```
model.predict(독립[:5])
```

```
import pandas as pd
iris = pd.read_csv('https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris2.csv')
# 원핫인코딩    df = pd.get_dummies(df)은 숫자 데이터이므로 실행 X
# =>> 품종 타입을 범주형으로 바꾸어 준다.
iris['품종'] = iris['품종'].astype('category')
iris = pd.get_dummies(iris)
iris
```

```
iris.isna().sum()
```

```
iris = iris.fillna(iris['꽃잎폭'].mean())
```

```
파일경로 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
boston = pd.read_csv(파일경로)
boston.columns
```

```
독립 = boston[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
종속 = boston[['medv']]
```

```
X = tf.keras.layers.Input(shape=[13])

H = tf.keras.layers.Dense(8)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

Y = tf.keras.layers.Dense(1)(H)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```
model.fit(독립, 종속, epochs=1000)
```
